```python
table = data.find('table',class_ ='ds-w-full ds-table ds-table-md ds-table-auto')
print(table)
```

```python
t=table.find_all('span',class_='ds-text-tight-s ds-font-medium ds-uppercase ds-text-left ds-text-typo')
teams=[]
for i in t:
    teams.append(i.text)
print(teams)
```
