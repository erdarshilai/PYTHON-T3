```python
import requests
from bs4 import BeautifulSoup
response=requests.get(r'https://books.toscrape.com/')
data=BeautifulSoup(response.text,'html.parser')
books=data.find('ol',class_='row')
t=books.find_all('a',{'title':True})
print(t)
```

```python
titles=[]
for i in t:
    book =i['title']
    titles.append(book)
print(titles)
```

```python
p=books.find_all('p',class_='price_color')
print(p)
```

```python
price=[]
for i in p:
    price.append(i.text)
print(price)
```

```python
image=books.find_all('img',class_='thumbnail')
print(image)
```

```python
img=[]
for i in image:
    img.append(i['src'])
print(img)
```

```python
import pandas as pd
d={
    'Title':titles,
    'Price':price,
    'Image Source':img
}
df=pd.DataFrame(d)
df
```

```python
response=requests.get(r'https://www.politifact.com/factchecks/list/')
```

```python
data=BeautifulSoup(response.text,'html.parser')
print(data.prettify())
```

```python
s=data.find_all('div',class_='m-statement__quote')
statement=[]
for i in s:
    statement.append(i.text.strip())
print(statement)
```

```python
m=data.find_all('div',class_='m-statement__meta')
source=[]
for i in m:
    source.append(i.find('a').text.strip())
print(source)
```

```python
d=data.find_all('footer',class_='m-statement__footer')
date=[]
for i in d:
    ans=i.text.split('•')[1]
    date.append(ans.strip())
print(date)
author=[]
for i in d:
    ans=i.text.split('•')[0]
    author.append(ans.strip('By'))
print(author)
```

```python
img=data.find_all('img',{'alt':True},class_='c-image__original')
result=[]
for i in img:
    result.append(i['alt'])
print(result)
```

```python
d={
    'Statement':statement,
    'Source':source,
    'Date':date,
    'Author':author,
    'Result':result
}
df=pd.DataFrame(d)
df
```

```python
import requests
from bs4 import BeautifulSoup
response=open('indeed jobs.html',encoding='utf-8')
data=BeautifulSoup(response,'html.parser')
print(data)
```

```python
jobs=data.find_all('div',class_='job_seen_beacon')
print(len(jobs))
```

```python
d={'name':[],'company':[],'Location':[],'Details':[]}
for i in jobs:
    n=i.find('h2',class_="jobTitle css-1psdjh5 eu4oa1w0")
    d['name'].append(n.text)
    k=i.find('div',class_="css-1afmp4o e37uo190")
    d['company'].append(k.text)
    m=i.find('div',class_="css-1restlb eu4oa1w0")
    d['Location'].append(m.text)
    o=i.find('div',class_="jobMetaDataGroup css-qspwa8 eu4oa1w0")
    if o:
        d['Details'].append(o.text)
    else:
        d['details'].append('')
df=pd.DataFrame(d)
df
#print(d)
```

```python
f=open('Scrape_this.html',encoding='utf-8')
data=BeautifulSoup(f,'html.parser')
```

```python
n=data.find_all('td',class_='')
print(len(n))
name=[]
for i in n:
    name.append(i.text)
print(name)
```

```python
t=data.find_all('td',class_='txt-right')
cols=[]
for i in t:
    cols.append(i.text)
cp=cols[0::5]
low=cols[1::5]
high=cols[2::5]
value=cols[3::5]
volume=cols[4::5]
print(cp)
print(low)
print(high)
print(value)
print(volume)
```

```python
c=data.find_all('td',class_="txt-ctr")
change=[]
for i in c:
    change.append(i.text)
print(change)
```

```python
d={
    'company Name':name,
    'current price':cp,
    'change':change,
    'Low':low,
    'High':high,
    'Value':value,
    'Volume':volume
}
df=pd.DataFrame(d)
df
```
